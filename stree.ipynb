```python
import pandas as pd
```

```python
df = pd.read_csv('StressLevelDataset.csv')
```

```python
df.head()
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['academic_performance', 'study_load', 'future_career_concerns',
       'extracurricular_activities'],inplace=True,axis=1)
```

```python
y = df['stress_level']
```

```python
df.drop('stress_level',axis=1,inplace=True)
```

```python
df.columns
```

```python
df.info()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier   # or use any other model
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df, 
    y, 
    test_size=0.2,      # 20% of data for testing (you can adjust this)
    random_state=42     # for reproducibility
)
```

```python
y
```

```python
# Initialize the model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train (fit) the model
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Detailed classification metrics
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
